- Crear un array de NumPy de dimensiones 5x5, con números enteros aleatorios en el rango [-40, 40]
- Hacer los siguientes reemplazos en orden
    - Reemplazar todos los valores multiplos de 3 por 1
    - Reemplazar todos los valores pares positivos por 0
    - Reemplazar todos los valores pares negativos por Nan
- Averiguar cuantos elementos tienen Nan por fila

```python

p1 = np.random.randint(-40,40,size=(5,5))


''' Reemplazar todos los valores multiplos de 3 por 1 '''
p1[p1%3==0]=1
print(p1)

''' Reemplazar todos los valores pares positivos por 0 '''

p1_2 =np.where(((p1%2==0) & (p1 > 0)) ,0,p1)

''' Reemplazar todos los valores pares negativos por Nan '''
p1_2 =np.where(((p1%2==0) & (p1 < 0)),np.nan,p1_2)
print(p1_2)
```

## 04 - 1.25pt
- 200 personas hacen una review de una película, dándole una puntuación de 0.0 a 5.0 a dicha película
- Crear una lista aleatoria con las posibles puntuaciones, es decir, 200 puntuaciones entre 0.0 y 5.0
- Discretizar dichas puntuaciones en 4 categorias [muy mala, mala, buena, excelente], los límites de cada categoría son a libre elección
- Discretizar dichas puntuaciones en 5 categorías distribuidas con rangos homogéneos (similar cantidad de valores en cada categoría)
- Mostrar las categorías creadas y cuantos elementos hay en cada una de ellas

```python
df = pd.DataFrame({'numero': np.random.randint(0, 5, 200)})
df['bins'] = pd.cut(x=df['numero'], bins=4,
                    labels=['muy mala', 'mala', 'buena',
                            'excelente'])

df['categoria'] = np.where(df['numero']<=1,"1",
                           np.where(df['numero']<=2,"2",
                            np.where(df['numero']<=3,"3",
                              np.where(df['numero']<=4,"4","5"))))

print(df)

df['categoria'].value_counts()
```

#### Preguntas

```python
from google.colab import files
uploaded = files.upload()
file_path = list(uploaded.keys())[0]
df = pd.read_csv(file_path,sep=';')

display(df.head())
```

```python
df['download_date'] = df['download_date'].str[0:9]

df['download_date'] = pd.to_datetime(df['download_date'], format='%Y-%m-%d')

li_fecha,ls_fecha = df['updated'].min(),df['updated'].max()
print(f"\n El rango de fechas contemplado en el cojunto de datos va desde {li_fecha} hasta {ls_fecha}  ")
```

```python
# Eliminar filas con valores nulos en la columna 'co'
datos_sin_nulos = df.dropna(subset=['station'])

# Calcular la media de ingreso para cada objectid
ingreso_por_objectid = datos_sin_nulos.groupby('station')['free'].mean().reset_index()

# Crear el gráfico de barras
fig = px.bar(ingreso_por_objectid, x='station', y='free', title='media más alta de bicis disponibles')

# Ajustes estéticos
fig.update_layout(xaxis_title='station', yaxis_title='media más alta de bicis disponibles', xaxis_tickangle=-45)

# Mostrar el gráfico
fig.show()

max_media_co = ingreso_por_objectid['free'].max()
estacion_max_media = ingreso_por_objectid.loc[ingreso_por_objectid['free'] == max_media_co, 'station'].iloc[0]

print(f'\n El número de estación con la media más alta de bicis disponibles fue: {estacion_max_media}')
```

```python
valores_histograma = list(df[df['station']== str(estacion_max_media)]['free'])

# Convertir la lista de datos en un DataFrame de pandas
df = pd.DataFrame(valores_histograma, columns=['valor'])

# Crear el histograma con Plotly
fig = px.histogram(df, x='valor', nbins=df['valor'].nunique(), title=f'histograma de dióxido de nitrógeno de la estación {estacion_max_media}')
fig.update_xaxes(title='Estación')
fig.update_yaxes(title='Frecuencia')

# Mostrar el histograma
fig.show()
```

```python
filtro = valores_histograma = df[df['station'] == str(estacion_max_media)][['download_date', 'free']]

# Convertir la columna de fecha_carga a tipo datetime si aún no está en ese formato
filtro['download_date'] = pd.to_datetime(filtro['download_date'])

# Establecer la columna de fecha_carga como el índice del DataFrame
filtro.set_index('download_date', inplace=True)

# Configurar el estilo de seaborn
sns.set_style("whitegrid")

# Graficar la serie de tiempo
plt.figure(figsize=(12, 6))
sns.lineplot(data=filtro, x=filtro.index, y='o3', color='skyblue', linewidth=2, markers='o', markerfacecolor='blue', markeredgewidth=0)

# Anotar los valores en cada punto
for index, row in filtro.iterrows():
    plt.annotate(f'{int(row["o3"])}', xy=(index, row["o3"]), xytext=(-10, 5), textcoords='offset points', fontsize=9)

plt.title('Serie de tiempo de bicis', fontsize=16)
plt.xlabel('Fecha', fontsize=14)
plt.ylabel('Valor de bicis', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.show()
```
